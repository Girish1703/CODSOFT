```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/sample_data/tested.csv")
print(df)

#TOTAL NUMBER OF FEMALE SURVIVING PASSENGER
female_surviving_passenger = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
total_female_surviving_passengers = len(female_surviving_passenger)
print("Total number of female surviving passengers:", total_female_surviving_passengers)

#TOTAL NUMBER OF MALE SURVIVING PASSENGER
male_surviving_passenger = df[(df['Sex'] == 'male') & (df['Survived'] == 0)]
total_male_surviving_passengers = len(male_surviving_passenger)
print("Total number of male surviving passengers:", total_male_surviving_passengers)

#MALE PASSENGER OF AGE ABOVE 60
male_above_60 = df[(df['Sex'] == 'male') & (df['Age'] > 50)]
total_male_above_60 = len(male_above_60)
print("Total number of male above 60 age:", total_male_above_60)

#FEMALE PASSENGER OF AGE ABOVE 40
female_above_40 = df[(df['Sex'] == 'male') & (df['Age'] > 40)]
total_female_above_40 = len(female_above_40)
print("Total number of female above 40 age:", total_female_above_40)

#PASSENGER IN 2 PCLASS
total_passengers_2Pclass = df[df['Pclass']==2]
total_passengers_2Pclass = len(total_passengers_2Pclass)
print("total_passengers_2Pclass:", total_passengers_2Pclass)

#TOTAL AMOUNT OF TICKET
total_ticket_amount = df['Ticket'].sum()
print("total_ticket_amount:",total_ticket_amount)

#SURVIVING FEMALE PASSENGER NAMELIST
survived_male_passenger = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
survived_male_passenger = survived_male_passenger['Name'].tolist()
print("survived_male_passenger:", survived_male_passenger)

#PASSENGER ID BELOW 900
passenger_id = df[df['PassengerId'] < 900]
passenger_id = len(passenger_id)
print("passenger_id:",passenger_id)

#TOTAL FEMALE PASSENGER DETAILS
female_passenger_details = df.loc[df['Sex'] == 'female', ['PassengerId', 'Survived', 'Pclass', 'Age', 'Name', 'Ticket']]
print("Details of female passengers:",female_passenger_details)

#PASSENGER IN S EMBARKED
total_passengers_SEmbarked = df[df['Embarked']=='S']
total_passengers_SEmbarked = len(total_passengers_SEmbarked)
print("total_passengers_SEmbarked:", total_passengers_SEmbarked)


#PIE CHART FOR SURVIVAL AND DEATH RATE
survived_counts = df['Survived'].value_counts()
labels = ['Did not Survive', 'Survived']
colors = ['red', 'green']
explode = (0.1, 0)
plt.pie(survived_counts, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Surviving vs Death Rate of Passengers')
plt.show()

#LINE CHART OF FEMALE AND MALE SURVIVING RATE
total_female_passengers = len(df[df['Sex'] == 'female'])
total_male_passengers = len(df[df['Sex'] == 'male'])
female_surviving_passenger = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
male_surviving_passenger = df[(df['Sex'] == 'male') & (df['Survived'] == 1)]
total_female_surviving_passengers = len(female_surviving_passenger)
total_male_surviving_passengers = len(male_surviving_passenger)
female_survival_rate = total_female_surviving_passengers / total_female_passengers
male_survival_rate = total_male_surviving_passengers / total_male_passengers
categories = ['Female', 'Male']
survival_rates = [female_survival_rate, male_survival_rate]
plt.plot(categories, survival_rates, marker='o')
plt.title('Surviving Rates of Female and Male Passengers')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.grid(True)
plt.show()
```
