```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv ("/content/sample_data/IRIS.csv")
print(df)

#TOTAL NO OF IRIS SETOSA SPECIES
setosa_count = df[df['species'] == 'Iris-setosa'].shape[0]
print("Total number of Iris-setosa:", setosa_count)

#TOTAL NO OF IRIS-VERSICOLOR SPECIES
versicolor_count = df[df['species'] == 'Iris-versicolor'].shape[0]
print("Total number of Iris-versicolor:",versicolor_count)

#TOTAL NO OF IRIS-VIRGINICA SPECIES
virginica_count = df[df['species'] == 'Iris-virginica'].shape[0]
print("Total number of Iris-virginica:",virginica_count)

#SETOSA SPECIES DETAILS
setosa_details = df.loc[df['species'] == 'Iris-setosa', ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
print("setosa-details:",setosa_details)

#VIRGINICA SPECIES DETAILS
virginica_details = df.loc[df['species'] == 'Iris-virginica', ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
print("virginica-details:",virginica_details)

#SEPAL LENGTH OF 4.9
sepal_length_count = df[df['sepal_length'] == 4.9]
sepal_length_count = len(sepal_length_count)
print("sepal_length_4.9_count:",sepal_length_count)

#PETAL LENGTH OF 1.4
petal_length_count = df[df['petal_length'] == 1.4]
petal_length_count = len(petal_length_count)
print("petal_length_1.4_count:",petal_length_count)

#HOW MANY IRIS-SETOSA OF 0.2 PETAL WIDTH
total_setosa= df[(df['petal_width'] == 0.2) & (df['species'] == 'Iris-setosa')]
total_setosa = len(total_setosa)
print("total_0.2_setosa:",total_setosa)

#SPECIES OF 5 SEPAL LENGTH
sepal_length_species = df[(df['sepal_length'] == 5) & (df['species'])]
print("5_sepal_length_species:", sepal_length_species)

#BAR CHART OF SPECIES
species_counts = df['species'].value_counts()
plt.bar(species_counts.index, species_counts.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Iris Species Distribution')
plt.show()

#PIE CHART OF SPECIES
species_counts = df['species'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Iris Species')
plt.show()
```
