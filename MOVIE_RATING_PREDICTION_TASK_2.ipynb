```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/content/sample_data/advertising.csv")
print(df)

#SUM OF TOTAL TV RATING
total_TV_rating = df['TV'].sum()
print("Total_TV_Rating:", total_TV_rating)

#MEAN OF TOTAL SALES RATING
total_mean_sales = df['Sales'].mean()
print("total_mean_sales:",total_mean_sales)

#TOTAL MOVIE RATING OF RADIO, NEWSPAPER, RADIO AND SALES
total_movie_rating = df[['TV', 'Radio', 'Newspaper', 'Sales']].sum().sum()
print("total_movie_rating:",total_movie_rating)

#HIGHEST RATING BETWEEN RADIO AND NEWSPAPER
highest_rating = max(df['Radio'].sum(), df['Newspaper'].sum())
print("Highest Rating:", highest_rating)

#IDENTIFYING THE MAXIMUM RATING
maximum_rating = df[['TV', 'Radio', 'Newspaper', 'Sales']].max().max
print("Maximum rating:", maximum_rating)

#PIE CHART OF MOVIE RATE PREDICTION
total_movie_rating = df[['TV', 'Radio', 'Newspaper', 'Sales']].sum().sum()
proportions = df[['TV', 'Radio', 'Newspaper', 'Sales']].sum() / total_movie_rating
labels = ['TV', 'Radio', 'Newspaper', 'Sales']
plt.pie(proportions, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Proportion of Movie Rating Types')
plt.show()

#LINE CHART OF MOVIE RATE PREDICTION
plt.figure(figsize=(10, 6))
plt.plot(df['TV'].head(10), label='TV')
plt.plot(df['Radio'].head(10), label='Radio')
plt.plot(df['Newspaper'].head(10), label='Newspaper')
plt.plot(df['Sales'].head(10), label='Sales')
plt.title('Movie Ratings Over Different Platforms (First 10 elements)')
plt.xlabel('Index')
plt.ylabel('Rating')
plt.legend()
plt.show()
```
